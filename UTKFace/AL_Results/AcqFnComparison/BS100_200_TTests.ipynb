```
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from google.colab import drive
drive.mount("/content/gdrive")
```

Here we conduct one-tailed t-tests comparing EI RMSE to benchmark RMSE for a given initial training set and active learning iteration


```
#Loading BS100 results
folder_path = "/content/gdrive/My Drive/FINAL_PAPER_ACTIVE_LEARNING_EXP/UTKFace/AL_Results_Start_150/"
total_itr = 2000//100 + 1

#Loading EI Results
ei_1 = np.sqrt(np.load(folder_path+"EITauInv5.5477MSEBS100_Ind1.npy")[0:total_itr])
ei_2 = np.sqrt(np.load(folder_path+"EITauInv5.5477MSEBS100_Ind2.npy")[0:total_itr])
ei_3 = np.sqrt(np.load(folder_path + "EITauInv5.5477MSEBS100_Ind3.npy")[0:total_itr])

#Loading Rand results
rand_1 = np.sqrt(np.load(folder_path + "RandMSEBS100_Ind1.npy")[0:total_itr])
rand_2 = np.sqrt(np.load(folder_path + "RandMSEBS100_Ind2.npy")[0:total_itr])
rand_3 = np.sqrt(np.load(folder_path + "RandMSEBS100_Ind3.npy")[0:total_itr])

#Loading MaxVar Results
maxvar_1 = np.sqrt(np.load(folder_path+"MaxVarMSEBS100_Ind1.npy")[0:total_itr])
maxvar_2 = np.sqrt(np.load(folder_path+"MaxVarMSEBS100_Ind2.npy")[0:total_itr])
maxvar_3 = np.sqrt(np.load(folder_path+"MaxVarMSEBS100_Ind3.npy")[0:total_itr])
```

```
#t test comparing EI RMSE to random RMSE for batch size 100
diff_1 = rand_1 - ei_1
diff_2 = rand_2 - ei_2
diff_3 = rand_3 - ei_3
popmean = 0.0
all_mse_diff = np.array(list(diff_1) + list(diff_2) + list(diff_3))
print(all_mse_diff.shape)
stats.ttest_1samp(all_mse_diff, popmean)
```

```
pval = 0.009592754644220054/2.0
print(pval)
```

```
#t test comparing EI RMSE to max. variance RMSE for batch size 100
diff_1 = maxvar_1 - ei_1
diff_2 = maxvar_2 - ei_2
diff_3 = maxvar_3 - ei_3
popmean = 0.0
all_mse_diff = np.array(list(diff_1) + list(diff_2) + list(diff_3))
print(all_mse_diff.shape)
stats.ttest_1samp(all_mse_diff, popmean)
```

```
pval = 6.199423143719838e-09/2.0
print(pval)
```
