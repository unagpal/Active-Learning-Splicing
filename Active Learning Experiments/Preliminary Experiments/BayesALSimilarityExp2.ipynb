```python
# Imports
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
```

```python
#Hamming Distance: Number of Positions with Different Bases
```

```python
#Experiment 1 Mean and Standard Deviation
rand_1 = np.load("/home/unagpal/mismatch_1_rand.npy")
bald_1 = np.load("/home/unagpal/mismatch_1_bald.npy")
mean_rand_1 = np.mean(rand_1)
mean_bald_1 = np.mean(bald_1)
std_rand_1 = np.std(rand_1)
std_bald_1 = np.std(bald_1)
print("Experiment 1: Mean # Mismatches Random Acquisition: " + str(mean_rand_1))
print("Experiment 1: Standard Deviation of # Mismatches Random Acquisition: " + str(std_rand_1))
print("Experiment 1: Mean # Mismatches BALD Acquisition: " + str(mean_bald_1))
print("Experiment 1: Standard Deviation of # Mismatches BALD Acquisition: " + str(std_bald_1))
```

```python
#Experiment 1 Graph
Queries = np.arange(0, len(rand_1), 1)
plt.figure(figsize=(12, 8), dpi=80)
plt.plot(Queries, rand_1, color="red", linewidth=1.0, marker='.', label="Dropout BALD")
plt.plot(Queries, bald_1, color="black", linewidth=1.0, marker='.', label="Dropout Random" )

plt.xlabel("Number of Acquired Samples from Pool Set (in addition to original 100 training points)")
plt.ylabel("Average # Mismatches of Acquired Point")
plt.title("Bayesian CNN Active Learning Acquisition Functions on 5' Rosenberg data")
plt.grid()

plt.legend(loc = 1)
plt.show()
```
