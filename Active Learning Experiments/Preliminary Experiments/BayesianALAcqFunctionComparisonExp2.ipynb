```python
# Imports
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
```

```python
bald_mse_2 = np.load("/home/unagpal/BaldMSEExp2.npy")
rand_mse_2 = np.load("/home/unagpal/RandMSEExp2.npy")
itr_2 = len(bald_mse_2)
bald_mse_1 = np.load("/home/unagpal/BaldMSEExp1.npy")[0:itr_2]
rand_mse_1 = np.load("/home/unagpal/RandMSEExp1.npy")[0:itr_2]
bald_mse_avg = (bald_mse_1 + bald_mse_2)/2.0
rand_mse_avg = (rand_mse_1 + rand_mse_2)/2.0
Queries = np.arange(0, itr_2, 1)
```

```python
#Graphing MSE for Experiment 1
plt.figure(figsize=(12, 8), dpi=80)
plt.plot(Queries, bald_mse_1, color="red", linewidth=1.0, marker='.', label="Dropout BALD")
plt.plot(Queries, rand_mse_1, color="black", linewidth=1.0, marker='.', label="Dropout Random" )

plt.xlabel("Number of Acquired Samples from Pool Set (in addition to original 100 training points)")
plt.ylabel("MSE on Test Set")
plt.title("Bayesian CNN Active Learning Acquisition Functions on 5' Rosenberg data")
plt.grid()

plt.legend(loc = 1)
plt.show()
```

```python
#Graphing MSE for Experiment 2
plt.figure(figsize=(12, 8), dpi=80)
plt.plot(Queries, bald_mse_2, color="red", linewidth=1.0, marker='.', label="Dropout BALD")
plt.plot(Queries, rand_mse_2, color="black", linewidth=1.0, marker='.', label="Dropout Random" )

plt.xlabel("Number of Acquired Samples from Pool Set (in addition to original 100 training points)")
plt.ylabel("MSE on Test Set")
plt.title("Bayesian CNN Active Learning Acquisition Functions on 5' Rosenberg data")
plt.grid()

plt.legend(loc = 1)
plt.show()
```

```python
#Graphing Average MSE Over Experiments
plt.figure(figsize=(12, 8), dpi=80)
plt.plot(Queries, bald_mse_avg, color="red", linewidth=1.0, marker='.', label="Dropout BALD")
plt.plot(Queries, rand_mse_avg, color="black", linewidth=1.0, marker='.', label="Dropout Random" )

plt.xlabel("Number of Acquired Samples from Pool Set (in addition to original 100 training points)")
plt.ylabel("Average MSE on Test Set")
plt.title("Bayesian CNN Active Learning Acquisition Functions on 5' Rosenberg data")
plt.grid()

plt.legend(loc = 1)
plt.show()
```
