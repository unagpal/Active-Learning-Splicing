```python
# Imports
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
```

```python
# Loading data
Dropout_Bald = np.load('BaldMSE.npy')
Dropout_Rand = np.load('RandMSE.npy')
Queries = np.arange(0, 101, 1)
Dropout_Bald_Avg = (Dropout_Bald[0] + Dropout_Bald[1])/2.0
Dropout_Rand_Avg = (Dropout_Rand[0] + Dropout_Rand[1])/2.0
```

```python
#Graphing MSE for Experiment 1
plt.figure(figsize=(12, 8), dpi=80)
plt.plot(Queries, Dropout_Bald[0], color="red", linewidth=1.0, marker='.', label="Dropout BALD")
plt.plot(Queries, Dropout_Rand[0], color="black", linewidth=1.0, marker='.', label="Dropout Random" )

plt.xlabel("Number of Acquired Samples from Pool Set (in addition to original 500 training points)")
plt.ylabel("MSE on Test Set")
plt.title("Bayesian CNN Active Learning Acquisition Functions on 5' Rosenberg data")
plt.grid()

plt.legend(loc = 4)
plt.show()
```

```python
#Graphing MSE for Experiment 2
plt.figure(figsize=(12, 8), dpi=80)
plt.plot(Queries, Dropout_Bald[1], color="red", linewidth=1.0, marker='.', label="Dropout BALD")
plt.plot(Queries, Dropout_Rand[1], color="black", linewidth=1.0, marker='.', label="Dropout Random" )

plt.xlabel("Number of Acquired Samples from Pool Set (in addition to original 500 training points)")
plt.ylabel("MSE on Test Set")
plt.title("Bayesian CNN Active Learning Acquisition Functions on 5' Rosenberg data")
plt.grid()

plt.legend(loc = 4)
plt.show()
```

```python
#Graphing Average MSE
plt.figure(figsize=(12, 8), dpi=80)
plt.plot(Queries, Dropout_Bald_Avg, color="red", linewidth=1.0, marker='.', label="Dropout BALD")
plt.plot(Queries, Dropout_Rand_Avg, color="black", linewidth=1.0, marker='.', label="Dropout Random" )

plt.xlabel("Number of Acquired Samples from Pool Set (in addition to original 500 training points)")
plt.ylabel("MSE on Test Set")
plt.title("Bayesian CNN Active Learning Acquisition Functions on 5' Rosenberg data")
plt.grid()

plt.legend(loc = 4)
plt.show()
```
