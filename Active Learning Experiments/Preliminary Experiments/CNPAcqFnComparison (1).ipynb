```python
# Imports
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
```

```python
#attached signifies all acquired points included in all context sets during training
maxvar_mse_2_attached = np.load("/home/unagpal/CNPMSEMaxVarExp2.npy")
rand_mse_2 = np.load("/home/unagpal/CNPMSERandExp2.npy")
maxvar_mse_2 = np.load("/home/unagpal/CNPMSEMaxVarModExp2.npy")
rand_mse_2_attached = np.load("/home/unagpal/CNPMSERandModExp2.npy")
itr_2_attached = min(len(maxvar_mse_2_attached), len(rand_mse_2_attached))
itr_2 = min(len(rand_mse_2), len(maxvar_mse_2))
maxvar_mse_1_attached = np.load("/home/unagpal/CNPMSEMaxVarExp1.npy")
rand_mse_1 = np.load("/home/unagpal/CNPMSERandExp1.npy")
maxvar_mse_1 = np.load("/home/unagpal/CNPMSEMaxVarModExp1.npy")
rand_mse_1_attached = np.load("/home/unagpal/CNPMSERandModExp1.npy")
itr_1_attached = min(len(maxvar_mse_1_attached), len(rand_mse_1_attached))
itr_1 = min(len(rand_mse_1), len(maxvar_mse_1))
itr_avg = min(itr_1, itr_2)
maxvar_mse_avg = (maxvar_mse_1[0:itr_avg] + maxvar_mse_2[0:itr_avg])/2.0
rand_mse_avg = (rand_mse_1[0:itr_avg] + rand_mse_2[0:itr_avg])/2.0
Queries_2 = np.arange(0, itr_2, 1)
Queries_2_attached = np.arange(0, itr_2_attached, 1)
Queries_1 = np.arange(0, itr_1, 1)
Queries_1_attached = np.arange(0, itr_1_attached, 1)
Queries_avg = np.arange(0, itr_avg, 1)
```

```python
#Graphing MSE for Experiment 1, where all acquired points are attached to all context sets
plt.figure(figsize=(12, 8), dpi=80)
plt.plot(Queries_1_attached, maxvar_mse_1_attached[0:itr_1_attached], color="red", linewidth=1.0, marker='.', label="MaxVar Acquisition")
plt.plot(Queries_1_attached, rand_mse_1_attached[0:itr_1_attached], color="black", linewidth=1.0, marker='.', label="Rand Acquisition" )

plt.xlabel("Number of Acquired Samples from Pool Set (in addition to original 100 training points)")
plt.ylabel("MSE on Test Set")
plt.title("CNP Active Learning Acquisition Functions on 5' Rosenberg data")
plt.grid()

plt.legend(loc = 1)
plt.show()
```

```python
#Graphing MSE for Experiment 1, where all context points are randomly selected from all the training data
plt.figure(figsize=(12, 8), dpi=80)
plt.plot(Queries_1, maxvar_mse_1[0:itr_1], color="red", linewidth=1.0, marker='.', label="MaxVar Acquisition")
plt.plot(Queries_1, rand_mse_1[0:itr_1], color="black", linewidth=1.0, marker='.', label="Rand Acquisition" )

plt.xlabel("Number of Acquired Samples from Pool Set (in addition to original 100 training points)")
plt.ylabel("MSE on Test Set")
plt.title("CNP Active Learning Acquisition Functions on 5' Rosenberg data")
plt.grid()

plt.legend(loc = 1)
plt.show()
```

```python
#Graphing MSE for Experiment 2, where all acquired points are attached to all context sets
plt.figure(figsize=(12, 8), dpi=80)
plt.plot(Queries_2_attached, maxvar_mse_2_attached[:itr_2_attached], color="red", linewidth=1.0, marker='.', label="MaxVar Acquisition")
plt.plot(Queries_2_attached, rand_mse_2_attached[:itr_2_attached], color="black", linewidth=1.0, marker='.', label="Rand Acquisition" )

plt.xlabel("Number of Acquired Samples from Pool Set (in addition to original 100 training points)")
plt.ylabel("MSE on Test Set")
plt.title("CNP Active Learning Acquisition Functions on 5' Rosenberg data")
plt.grid()

plt.legend(loc = 1)
plt.show()
```

```python
#Graphing MSE for Experiment 2, where all context points are randomly selected from all the training data
plt.figure(figsize=(12, 8), dpi=80)
plt.plot(Queries_2, maxvar_mse_2[0:itr_2], color="red", linewidth=1.0, marker='.', label="MaxVar Acquisition")
plt.plot(Queries_2, rand_mse_2[0:itr_2], color="black", linewidth=1.0, marker='.', label="Rand Acquisition" )

plt.xlabel("Number of Acquired Samples from Pool Set (in addition to original 100 training points)")
plt.ylabel("MSE on Test Set")
plt.title("CNP Active Learning Acquisition Functions on 5' Rosenberg data")
plt.grid()

plt.legend(loc = 1)
plt.show()
```

```python
#Graphing Average MSE Over Both Experiments (where context points are sampled ranodmly)
plt.figure(figsize=(12, 8), dpi=80)
plt.plot(Queries_avg, maxvar_mse_avg, color="red", linewidth=1.0, marker='.', label="MaxVar Acquisition")
plt.plot(Queries_avg, rand_mse_avg, color="black", linewidth=1.0, marker='.', label="Rand Acquisition" )

plt.xlabel("Number of Acquired Samples from Pool Set (in addition to original 100 training points)")
plt.ylabel("MSE on Test Set")
plt.title("CNP Active Learning Acquisition Functions on 5' Rosenberg data")
plt.grid()

plt.legend(loc = 1)
plt.show()
```
