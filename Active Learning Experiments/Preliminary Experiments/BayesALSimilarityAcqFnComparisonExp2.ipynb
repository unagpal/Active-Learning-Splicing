```python
# Imports
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
```

```python
#Comparing Effectiveness of Variance-Similarity Acquisition Functions: Grid Search 1 over positive lambda
#Loading MSE Arrays
min_itr = 256
setting_1 = np.load("/home/unagpal/ModBaldMSEExp1_Setting1.npy")[:min_itr]
setting_2 =  np.load("/home/unagpal/ModBaldMSEExp1_Setting2.npy")[:min_itr]
setting_3 =  np.load("/home/unagpal/ModBaldMSEExp1_Setting3.npy")[:min_itr]
setting_4 =  np.load("/home/unagpal/ModBaldMSEExp1_Setting4.npy")[:min_itr]
setting_5 =  np.load("/home/unagpal/ModBaldMSEExp1_Setting5.npy")[:min_itr]
setting_6 =  np.load("/home/unagpal/ModBaldMSEExp1_Setting6.npy")[:min_itr]
setting_7 =  np.load("/home/unagpal/ModBaldMSEExp1_Setting7.npy")[:min_itr]
setting_8 =  np.load("/home/unagpal/ModBaldMSEExp1_Setting8.npy")[:min_itr]
setting_9 =  np.load("/home/unagpal/ModBaldMSEExp1_Setting9.npy")[:min_itr]
setting_10 =  np.load("/home/unagpal/ModBaldMSEExp1_Setting10.npy")[:min_itr]
```

```python
#Graphing MSE as function of number of acquired points
Queries = np.arange(0, min_itr, 1)
plt.figure(figsize=(12, 8), dpi=80)
plt.plot(Queries, setting_1, color="blue", linewidth=1.0, marker='.', label="Setting 1")
plt.plot(Queries, setting_2, color="green", linewidth=1.0, marker='.', label="Setting 2" )
plt.plot(Queries, setting_3, color="red", linewidth=1.0, marker='.', label="Setting 3" )
plt.plot(Queries, setting_4, color="cyan", linewidth=1.0, marker='.', label="Setting 4" )
plt.plot(Queries, setting_5, color="magenta", linewidth=1.0, marker='.', label="Setting 5" )
plt.plot(Queries, setting_6, color="yellow", linewidth=1.0, marker='.', label="Setting 6" )
plt.plot(Queries, setting_7, color="black", linewidth=1.0, marker='.', label="Setting 7" )
plt.plot(Queries, setting_8, color='0.7', linewidth=1.0, marker='.', label="Setting 8" )
plt.plot(Queries, setting_9, color="burlywood", linewidth=1.0, marker='.', label="Setting 9" )
plt.plot(Queries, setting_10, color="chartreuse", linewidth=1.0, marker='.', label="Setting 10" )
plt.xlabel("Number of Acquired Samples from Pool Set (in addition to original 100 training points)")
plt.ylabel("MSE on Test Data")
plt.title("Bayesian CNN BALD-Similarity Acquisition Functions on 5' Rosenberg data")
plt.grid()
plt.legend(loc = 1)
plt.show()
```
