```python
# Imports
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
```

```python
ei_mse_2 = np.load("/home/unagpal/EIMSEExp2.npy")
rand_mse_2 = np.load("/home/unagpal/RandMSEEIExp2.npy")
itr_2 = len(ei_mse_2)
ei_mse_1 = np.load("/home/unagpal/EIMSEExp1.npy")
rand_mse_1 = np.load("/home/unagpal/RandMSEEIExp1.npy")
itr_1 = len(ei_mse_1)
ei_mse_avg = (ei_mse_1[:itr_2] + ei_mse_2)/2.0
rand_mse_avg = (rand_mse_1[:itr_2] + rand_mse_2)/2.0
Queries = np.arange(0, itr_2, 1)
```

```python
#Graphing MSE for Experiment 1, on train/test indices 3
Queries_1 = np.arange(itr_1)
plt.figure(figsize=(12, 8), dpi=80)
plt.plot(Queries_1, ei_mse_1, color="red", linewidth=1.0, marker='.', label="Dropout EI")
plt.plot(Queries_1, rand_mse_1, color="black", linewidth=1.0, marker='.', label="Dropout Random" )

plt.xlabel("Number of Acquired Samples from Pool Set (in addition to original 100 training points)")
plt.ylabel("MSE on Test Set")
plt.title("Bayesian CNN Active Learning Acquisition Functions on 5' Rosenberg data")
plt.grid()

plt.legend(loc = 1)
plt.show()
```
