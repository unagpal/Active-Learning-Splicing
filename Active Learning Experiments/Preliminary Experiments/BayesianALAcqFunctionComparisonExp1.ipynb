```
from google.colab import drive
drive.mount('/content/gdrive')

# Imports
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
```

```
# Pool Sample Size for BALD is 5000
# Loading data
Dropout_Bald = np.load('/content/gdrive/My Drive/Active Learning Research 2019/BaldMSEExp1.npy')
Dropout_Rand = np.load('/content/gdrive/My Drive/Active Learning Research 2019/RandMSEExp1.npy')
Queries = np.arange(0, 101, 1)
Dropout_Bald_Avg = (Dropout_Bald[0] + Dropout_Bald[1]+Dropout_Bald[2])/3.0
Dropout_Rand_Avg = (Dropout_Rand[0] + Dropout_Rand[1] + Dropout_Rand[2])/3.0
```

```
#Graphing MSE for Experiment 1
plt.figure(figsize=(12, 8), dpi=80)
plt.plot(Queries, Dropout_Bald[0], color="red", linewidth=1.0, marker='.', label="Dropout BALD")
plt.plot(Queries, Dropout_Rand[0], color="black", linewidth=1.0, marker='.', label="Dropout Random" )

plt.xlabel("Number of Acquired Samples from Pool Set (in addition to original 400 training points)")
plt.ylabel("MSE on Test Set")
plt.title("Bayesian CNN Active Learning Acquisition Functions on 5' Rosenberg data")
plt.grid()

plt.legend(loc = 4)
plt.show()
```

```
#Graphing MSE for Experiment 2
plt.figure(figsize=(12, 8), dpi=80)
plt.plot(Queries, Dropout_Bald[1], color="red", linewidth=1.0, marker='.', label="Dropout BALD")
plt.plot(Queries, Dropout_Rand[1], color="black", linewidth=1.0, marker='.', label="Dropout Random" )

plt.xlabel("Number of Acquired Samples from Pool Set (in addition to original 400 training points)")
plt.ylabel("MSE on Test Set")
plt.title("Bayesian CNN Active Learning Acquisition Functions on 5' Rosenberg data")
plt.grid()

plt.legend(loc = 4)
plt.show()
```
