```python
# Imports
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
```

```python
# Loading data
MaxVar_MSE = np.load('CNP_MLP_MaxVar_MSE.npy')
Rand_MSE = np.load('CNP_MLP_Rand_MSE.npy')
Queries = np.arange(0, 101, 1)
Avg_MaxVar_MSE = (MaxVar_MSE[0,:] + MaxVar_MSE[1,:])/2.0
Avg_Rand_MSE = (Rand_MSE[0,:] + Rand_MSE[1,:])/2.0
```

```python
#Graphing MSE for Experiment 1
plt.figure(figsize=(12, 8), dpi=80)
plt.plot(Queries, MaxVar_MSE[0,:], color="red", linewidth=1.0, marker='.', label="MaxVar CNP")
plt.plot(Queries, Rand_MSE[0,:], color="black", linewidth=1.0, marker='.', label="Random CNP" )

plt.xlabel("Number of Acquired Samples from Pool Set (in addition to original 1000 training points)")
plt.ylabel("MSE on Test Set")
plt.title("Comparing CNP (MLP Encoder) Acquisition Functions on 5' Rosenberg data: Experiment 1")
plt.grid()

plt.legend(loc = 4)
plt.show()
```

```python
#Graphing MSE for Experiment 2
plt.figure(figsize=(12, 8), dpi=80)
plt.plot(Queries, MaxVar_MSE[1,:], color="red", linewidth=1.0, marker='.', label="MaxVar CNP")
plt.plot(Queries, Rand_MSE[1,:], color="black", linewidth=1.0, marker='.', label="Random CNP" )

plt.xlabel("Number of Acquired Samples from Pool Set (in addition to original 1000 training points)")
plt.ylabel("MSE on Test Set")
plt.title("Comparing CNP (MLP Encoder) Acquisition Functions on 5' Rosenberg data: Experiment 2")
plt.grid()

plt.legend(loc = 4)
plt.show()
```

```python
#Graphing MSE for Average of 2 Experiments
plt.figure(figsize=(12, 8), dpi=80)
plt.plot(Queries, Avg_MaxVar_MSE, color="red", linewidth=1.0, marker='.', label="MaxVar CNP")
plt.plot(Queries, Avg_Rand_MSE, color="black", linewidth=1.0, marker='.', label="Random CNP" )

plt.xlabel("Number of Acquired Samples from Pool Set (in addition to original 1000 training points)")
plt.ylabel("MSE on Test Set")
plt.title("Comparing CNP (MLP Encoder) Acquisition Functions on 5' Rosenberg data: Average MSE")
plt.grid()

plt.legend(loc = 4)
plt.show()
```
