```python
# Imports
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
```

```python
maxvar_mse_2 = np.load("/home/unagpal/CNPMSEMaxVarExp2.npy")
rand_mse_2 = np.load("/home/unagpal/CNPMSERandExp2.npy")
itr_2 = len(maxvar_mse_2)
maxvar_mse_1 = np.load("/home/unagpal/CNPMSEMaxVarExp1.npy")
rand_mse_1 = np.load("/home/unagpal/CNPMSERandExp1.npy")
maxvar_mse_avg = (maxvar_mse_1[:itr_2] + maxvar_mse_2)/2.0
rand_mse_avg = (rand_mse_1[:itr_2] + rand_mse_2)/2.0
Queries_2 = np.arange(0, itr_2, 1)
Queries_1 = np.arange(0, len(rand_mse_1), 1)
```

```python
#Graphing MSE for Experiment 1
plt.figure(figsize=(12, 8), dpi=80)
plt.plot(Queries_1, maxvar_mse_1, color="red", linewidth=1.0, marker='.', label="MaxVar Acquisition")
plt.plot(Queries_1, rand_mse_1, color="black", linewidth=1.0, marker='.', label="Rand Acquisition" )

plt.xlabel("Number of Acquired Samples from Pool Set (in addition to original 100 training points)")
plt.ylabel("MSE on Test Set")
plt.title("CNP Active Learning Acquisition Functions on 5' Rosenberg data")
plt.grid()

plt.legend(loc = 1)
plt.show()
```

```python
#Graphing MSE for Experiment 2
plt.figure(figsize=(12, 8), dpi=80)
plt.plot(Queries_2, maxvar_mse_2, color="red", linewidth=1.0, marker='.', label="MaxVar Acquisition")
plt.plot(Queries_2, rand_mse_2, color="black", linewidth=1.0, marker='.', label="Rand Acquisition" )

plt.xlabel("Number of Acquired Samples from Pool Set (in addition to original 100 training points)")
plt.ylabel("MSE on Test Set")
plt.title("CNP Active Learning Acquisition Functions on 5' Rosenberg data")
plt.grid()

plt.legend(loc = 1)
plt.show()
```

```python
#Graphing Average MSE Over Both Experiments
plt.figure(figsize=(12, 8), dpi=80)
plt.plot(Queries_2, maxvar_mse_avg, color="red", linewidth=1.0, marker='.', label="MaxVar Acquisition")
plt.plot(Queries_2, rand_mse_avg, color="black", linewidth=1.0, marker='.', label="Rand Acquisition" )

plt.xlabel("Number of Acquired Samples from Pool Set (in addition to original 100 training points)")
plt.ylabel("MSE on Test Set")
plt.title("CNP Active Learning Acquisition Functions on 5' Rosenberg data")
plt.grid()

plt.legend(loc = 1)
plt.show()
```
