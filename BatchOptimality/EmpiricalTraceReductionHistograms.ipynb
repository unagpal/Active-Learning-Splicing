```
import numpy as np
import matplotlib.pyplot as plt
import itertools

from google.colab import drive
drive.mount("/content/gdrive", force_remount=True)
```

```
#dir_path is set to the location containing empirical optimality ratios from which histograms are produced
#here it is set to the file path of the "EmpiricalOptimalityRatios" directory within "BatchOptimality"
dir_path = "/content/gdrive/My Drive/Active_Learning_Expected_Improvement_Maximization/BatchOptimality/EmpiricalOptimalityRatios/"

#Matplotlib font size settings for figure readibility (taken from https://stackoverflow.com/questions/3899980/how-to-change-the-font-size-on-a-matplotlib-plot)
SMALL_SIZE = 14
MEDIUM_SIZE = 18
BIGGER_SIZE = 18

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=BIGGER_SIZE)    # legend fontsize

def opt_ratio_histogram(fpath):
  """Takes in the file path containing the ratios of greedy trace reduction/optimal trace reduction for a specific experiment and displays a histogram
  #Arguments
    fpath: Additional file path after dir_path for ratios of greedy trace reduction/optimal trace reduction for the given experiment
  #Returns
    N/A, histogram should be shown
  """
  opt_ratios = np.load(dir_path + fpath)
  print("Min opt ratio: " + str(np.amin(opt_ratios)))
  plt.figure(figsize=(6,9), dpi=100)
  num_bins = 7
  n, bins, patches = plt.hist(opt_ratios, num_bins, facecolor='blue', alpha=0.5)
  plt.xlabel("Greedy Trace Reduction/Optimal Trace Reduction")
  plt.ylabel("Frequency")
  plt.show()
```

```
#Histograms illustrated for various batch sizes and number of iid Gaussian samples serving as a proxy for model predictions per point
opt_ratio_histogram("BS2Samp3OptRatios.npy")
```
